```python
# Library to read PDF files
from pdfminer.high_level import extract_text
# Read the PDF file and convert it to a string
pdf = extract_text("../../www/kardex/170300075.pdf")
```

```python
strings = pdf.split("\n\n")
```

```python
general_average = strings[-3]
total_credits = strings[-4]
```

```python
strings = strings[14:(len(strings)-7)]
strings
```

```python
stopwords = ['\x0cMatricula:',
 '170300075',
 'Nombre:',
 'DIAZ GONZALEZ / KENNETH',
 'Plan de Estudios:',
 '2016ID',
 'Situación:',
 'Regular',
 'KARDEX GENERAL',
 'OP1']
```

```python
new_strings = [s for s in strings if s not in stopwords]
```

```python
new_strings
```

```python
import re
```

```python
clave = []
asignatura = []
periodo = []

for s in new_strings:
    key = re.findall(r"[A-Z]{2}[0-9]{4}", s)
    subject = re.findall(r"[A-Z]{1}[a-z].+", s, re.UNICODE)
    period = re.findall(r"[0-9]{6}", s)

    if len(subject) > 0:
        asignatura.append(subject[0])
    if len(period) > 0:
        periodo.append(period[0])
    if len(key) > 0:
        clave.append(key[0])
```

```python
data = {"clave":clave, "asignatura":asignatura}
data
```
