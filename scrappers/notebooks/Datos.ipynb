```python
# Reemplazar una llave de un diccionario
# por un false
def replace_nan(dictionary, key):
    dictionary[key] = False
```

```python
# Encontrar y reemplazar valores nan por False
# en el diccionario
def find_nan(dictionary):
    for k, v in dictionary.items():
        if type(v) == dict:
            find_nan(v)

        else:
            if pd.isnull(v):
                dictionary[k] = False
```
